```python
import random

suits = ('Hearts', 'Diamonds', 'Spades', 'Clubs')
ranks = ('Two', 'Three', 'Four', 'Five', 'Six', 'Seven', 'Eight', 'Nine', 'Ten', 'Jack', 'Queen', 'King', 'Ace')
values = {'Two':2, 'Three':3, 'Four':4, 'Five':5, 'Six':6, 'Seven':7, 'Eight':8, 
            'Nine':9, 'Ten':10, 'Jack':10, 'Queen':10, 'King':10, 'Ace':11}

class Card():
    def __init__ (self,suit,rank):
        self.suit = suit
        self.rank = rank
        
    def __str__(self):
        return '{} of {}'.format(self.rank,self.suit)

class Deck():
    def __init__(self):
        self.deck = []
        for suit in suits:
            for rank in ranks:
                self.deck.append(Card(suit,rank))
    
    def shuffle(self):
        random.shuffle(self.deck)
    
    def deal_one(self):
        return self.deck.pop()
    
class Player():
    def __init__ (self,name,player_hand):
        self.name = name
        self.player_hand = player_hand
        
    def add_card(self,next_card):
        self.player_hand.append(next_card)
    
    def calculate_value(self,player_hand):
        for card in player_hand:
            return card.rank
    
class Chips():
    def __init__(self,bank=100):
        
        self.bank = bank
    
    def __str__(self):
        return str(self.bank)
        
    def lose_bet(self,bet):
        self.bank -= bet
    
    def win_bet(self,bet):
        self.bank += bet
    
def show_all(dealer,player):
    print('\n Dealer Hand')
    for card in dealer_hand:
        print(card)
        
    print('Player Hand')
    for card in player_hand:
        print(card)
        
def show_some(dealer,player):
    print('\n Dealer Hand \n')
    print('Card Hidden')
    print(dealer_hand[-1])
        
    print('\n Player Hand \n')
    for card in player_hand:
        print(card)
        
def check_win(dealer_total,player_total):
    if player_total>21:
        winner = 'dealer'   
    else:
        if dealer_total>21:
            winner = 'player'
        else:
            if dealer_total > player_total:
                winner = 'dealer'
            else:
                if dealer_total < player_total:
                    winner = 'player'
            
                else:
                    winner = 'draw'
    return winner

def calculate_player_value(player_hand):
    player_values = []
    
    for card in player_hand:
        player_values.append(values[card.rank])
    
    if "Ace" in player_hand:
        player_total = sum(player_values)-10
    else:
        player_total = sum(player_values)
    
    return player_total

    
def calculate_dealer_value(dealer_hand):
    dealer_values=[]    
    
    for card in dealer_hand:
        dealer_values.append(values[card.rank])
        
    if "Ace" in dealer_hand:
        dealer_total = sum(dealer_values)-10
    else:
        dealer_total = sum(dealer_values)
    
    return dealer_total

    
```
